Now i have a "Brand Channel" column if you remember. There is two option yes or no. I want to viseualize it.

```python
# Create a count plot for the "Brand Channel" column
sns.countplot(x="Brand channel", data=df)

# Add title and labels
plt.title("Distribution of Brand Channel")
plt.xlabel("Brand Channel")
plt.ylabel("Count")

# Show the plot
plt.show()
```

Make it different shape and colors. Maybe a cake chart or likesomething.

```python
# Calculate the count of each category
brand_channel_counts = df['Brand channel'].value_counts()

# Plot a pie chart
plt.figure(figsize=(5, 5))
plt.pie(brand_channel_counts, labels=brand_channel_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightgreen'])

# Add title
plt.title('Distribution of Brand Channel')

# Show the plot
plt.show()
```

We have "Subscribers (millions)" column dtype float. Now i want visualize it like two different group. I need one chart and two group. One is lower than 75 the other one is bigger than 75.



```python
# Define the groups based on subscribers count
df['Group'] = df['Subscribers (millions)'].apply(lambda x: 'Lower than 75M' if x < 75 else 'Higher than 75M')

# Count the number of channels in each group
group_counts = df['Group'].value_counts()

# Plot the bar chart
plt.figure(figsize=(8, 6))
group_counts.plot(kind='bar', color=['skyblue', 'lightgreen'])

# Add title and labels
plt.title('Subscribers Count Grouped by 75M Threshold')
plt.xlabel('Subscribers Count')
plt.ylabel('Number of Channels')

# Show the plot
plt.show()
```

Now for my other columns "Primary Language","Country","Category". I want different shape of charts. One by one i want to see my data crystal clear.

```python
# Count the number of channels for each language
language_counts = df['Primary language'].value_counts()

# Plot the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=language_counts.index, y=language_counts.values, palette="viridis")

# Add title and labels
plt.title('Number of Channels by Primary Language')
plt.xlabel('Primary Language')
plt.ylabel('Number of Channels')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
# Plot the count plot for the "Country" column
plt.figure(figsize=(12, 6))
sns.countplot(x='Country', data=df, palette="coolwarm")

# Add title and labels
plt.title('Distribution of Channels by Country')
plt.xlabel('Country')
plt.ylabel('Number of Channels')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
# Count the number of channels for each category
category_counts = df['Category'].value_counts()

# Plot the pie chart
plt.figure(figsize=(5, 5))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', colors=plt.cm.tab20.colors)

# Add title
plt.title('Distribution of Channels by Category')

# Show the plot
plt.show()
```

Now i need two different value. I want to check "Brand channel" according to "Subscribers (millions)" in one chart.

```python
# Set up the bar plot
plt.figure(figsize=(12, 8))

# Plot the bar plot
sns.barplot(data=df, x='Category', y='Subscribers (millions)', hue='Brand channel', palette=['skyblue', 'lightgreen'],errorbar=None)

# Add title and labels
plt.title('Subscribers by Category and Brand Channel')
plt.xlabel('Category')
plt.ylabel('Subscribers (millions)')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

Now i need another chart about top ten. X  = Name Y = Subscribers (millions) and i need if it is Brand channel.
I changed what you give y and x. Now "Name" is my x. But the text is not okey. I have to rotate that text. How can i do that.

```python
# Sort the dataframe by subscribers in descending order and select the top ten rows
top_ten = df.nlargest(10, 'Subscribers (millions)')

# Set up the bar plot
plt.figure(figsize=(10, 6))

# Plot the bar plot
ax = sns.barplot(data=top_ten, x='Name', y='Subscribers (millions)', hue='Brand channel', palette=['skyblue', 'lightgreen'])

# Rotate x-axis labels for better readability
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')

# Add title and labels
plt.title('Top Ten YouTube Channels by Subscribers')
plt.xlabel('Channel Name')
plt.ylabel('Subscribers (millions)')

# Show the plot
plt.show()
```
